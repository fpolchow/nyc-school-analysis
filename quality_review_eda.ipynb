```python
import pandas as pd
qual_review = pd.read_csv('./data/2005_-_2017_Quality_Review_Ratings.csv')
qual_review.head()
```

```python
## first look and see what the unique years are for quality review data
qual_review['SchYR'].unique()
```

```python
## mapping the quality review ratings to numbers
quality_dict = {'U':0,'UD':1,'D':2,'P':3,'WD':4}
numerical_quality_review = qual_review.replace(quality_dict)
numerical_quality_review.head()
```

```python
## remove non-null values
numerical_quality_review[qual_review['SchYR'] == 'SY16-17'].notnull().all()

## only keep these questions as they are the most recent. questions
columns_to_keep = ['BN','SchYR','QS1_1','QS1_2','QS1_3','QS1_4','QS2_2','QS3_1','QS3_4','QS4_1','QS4_2','QS5_1']
numerical_quality_review = numerical_quality_review[columns_to_keep]

## limit the years
year_list = ['SY11-12','SY12-13', 'SY13-14', 'SY15-16', 'SY16-17']

final_quality = numerical_quality_review.query('SchYR == @year_list')
```

```python
## save cleaned quality review
final_quality.to_csv('./cleaned_data/quality_review_cleaned.csv')
```
