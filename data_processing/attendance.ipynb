```python
def make_relative_absences(row):
    present_percentage = row['Present'] / (row['Present'] + row['Absent'])
    return present_percentage
```

```python
def make_relative_release(row):
    percent_released = row['Released'] / (row['Present'] + row['Absent'])
    return percent_released
```

```python
school_year_attendance['percent_present'] = school_year_attendance.apply(make_relative_absences,axis=1)
```

```python
school_year_attendance['relative_release'] = school_year_attendance.apply(make_relative_release,axis=1)
```
